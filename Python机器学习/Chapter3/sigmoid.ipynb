```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(z):
    return 1.0/(1.0+np.exp(-z))
```

```python
z = np.arange(-7,7,0.1)
phi_z = sigmoid(z)
plt.plot(z,phi_z)
plt.axvline(0.0,color='k')
plt.axhline(y=0.0,ls='dotted',color='k')
plt.axhline(y=0.5,ls='dotted',color='k')
plt.axhline(y=1.0,ls='dotted',color='k')
plt.yticks([0.0,0.5,1.0])
plt.ylim(-0.1,1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')
plt.show()
```

```python
x=np.linspace(1e-8,1-1e-8,1000)  #这个表示在-5到5之间生成1000个x值

y1 = -np.log(x)#对上述生成的1000个数循环用sigmoid公式求对应的y
y2 = -np.log(1-x)
plt.xlim(0,1)
plt.ylim(0,5)
plt.xlabel('$\phi$(z)')
plt.ylabel('$J(w)$')
plt.plot(x,y1,label = 'y = 1')
plt.plot(x,y2,label = 'y = 0',linestyle='--')
plt.legend()
plt.show()
```
